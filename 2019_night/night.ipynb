```python
level = 'ADM3_EN'
level2 = 'ADM2_EN'
nl1909 = pd.read_csv("201909_adm3.csv")[[level,level2,'layer']]
nl1910 = pd.read_csv("201910_adm3.csv")[[level,level2,'layer']]
nl1911 = pd.read_csv("201911_adm3.csv")[[level,level2,'layer']]
nl1912 = pd.read_csv("201912_adm3.csv")[[level,level2,'layer']]
nl2001 = pd.read_csv("202001_adm3.csv")[[level,level2,'layer']]
nl2002 = pd.read_csv("202002_adm3.csv")[[level,level2,'layer']]
nl2007 = pd.read_csv("202007_adm3.csv")[[level,level2,'layer']]
nl2008 = pd.read_csv("202008_adm3.csv")[[level,level2,'layer']]
nl2009 = pd.read_csv("202009_adm3.csv")[[level,level2,'layer']]
nl2010 = pd.read_csv("202010_adm3.csv")[[level,level2,'layer']]
nl2011 = pd.read_csv("202011_adm3.csv")[[level,level2,'layer']]
nl2012 = pd.read_csv("202012_adm3.csv")[[level,level2,'layer']]
nl1909[['year','month']] = 2019,9
nl1910[['year','month']] = 2019,10
nl1911[['year','month']] = 2019,11
nl1912[['year','month']] = 2019,12
nl2001[['year','month']] = 2020,1
nl2002[['year','month']] = 2020,2
nl2007[['year','month']] = 2020,7
nl2008[['year','month']] = 2020,8
nl2009[['year','month']] = 2020,9
nl2010[['year','month']] = 2020,10
nl2011[['year','month']] = 2020,11
nl2012[['year','month']] = 2020,12
nl1909['County'] = nl1909[level].str.lower()
nl1910['County'] = nl1910[level].str.lower()
nl1911['County'] = nl1911[level].str.lower()
nl1912['County'] = nl1912[level].str.lower()
nl2001['County'] = nl2001[level].str.lower()
nl2002['County'] = nl2002[level].str.lower()
nl2007['County'] = nl2007[level].str.lower()
nl2008['County'] = nl2008[level].str.lower()
nl2009['County'] = nl2009[level].str.lower()
nl2010['County'] = nl2010[level].str.lower()
nl2011['County'] = nl2011[level].str.lower()
nl2012['County'] = nl2012[level].str.lower()
nl1909['District'] = nl1909[level2].str.lower()
nl1910['District'] = nl1910[level2].str.lower()
nl1911['District'] = nl1911[level2].str.lower()
nl1912['District'] = nl1912[level2].str.lower()
nl2001['District'] = nl2001[level2].str.lower()
nl2002['District'] = nl2002[level2].str.lower()
nl2007['District'] = nl2007[level2].str.lower()
nl2008['District'] = nl2008[level2].str.lower()
nl2009['District'] = nl2009[level2].str.lower()
nl2010['District'] = nl2010[level2].str.lower()
nl2011['District'] = nl2011[level2].str.lower()
nl2012['District'] = nl2012[level2].str.lower()
lsts = [nl1909,nl1910,nl1911,nl1912,nl2001,nl2002,nl2007,nl2008,nl2009,nl2010,nl2011,nl2012]
# drop duplicated county name with different id (they have the same county and district name, but are identified as 
# different polygon in NL Extraction). Here I use the mean of two same counties. 
for j in lsts: 
    tr = j[j[['County','District']].duplicated(keep=False)].groupby('County').mean().reset_index().copy()
    for i in list(tr['County']):
        sh = j.loc[j['County']==i,'layer'].shape[0]
        j.loc[j['County']==i,'layer'] = tr[tr['County']==i]['layer'].values.repeat(sh)

GSEC1_nightlight.rename(columns={"layer":"NL_District"},inplace=True)
res = pd.concat([nl1909,nl1910],axis=0)
lst = [nl1911,nl1912,nl2001,nl2002,nl2007,nl2008,nl2009,nl2010,nl2011,nl2012]
for i in lst: 
    res = pd.concat([res,i],axis=0)
GSEC1_nightlight['County'] = GSEC1_nightlight['s1aq4a'].str.lower()
# change the unmatched county name 
res['County'].replace({'west budama (kisolo)':'west budama','kinkizi':'kinkiizi',
                         'bushenyi  municipality':'bushenyi-ishaka municipal','makindye ssabagabo municipality':'makindye ssabagabo munici',
                         'chekwi':'chekwii'},inplace=True)
GSEC1_output = pd.merge(GSEC1_nightlight,res[['layer','year','month','County','District']],how='left',on=['County','District','year','month'])
# drop duplicated cases 
GSEC1_output.drop_duplicates('hhid',inplace=True)
GSEC1_output.rename(columns={"layer":"NL_County"},inplace=True)
GSEC1_output.drop(columns=['ADM2_EN'],inplace=True) 
# drop those without date data 
GSEC1_output.dropna(subset=['year'],inplace=True)
# imputation 
s1 = GSEC1_output[GSEC1_output['NL_County'].isna()].copy()
GSEC1_output.loc[GSEC1_output['NL_County'].isna(),'NL_County'] = s1['NL_District']
GSEC1_output.to_csv("GSEC1_nl_dist.csv",index=False) 
```

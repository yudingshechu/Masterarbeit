```python
import pandas as pd
import numpy as np
import os
import re

def NDVIextract(motherpath, return_name ):
    adms = os.listdir(motherpath)
    pattern = r'- (.*?)__'  # Regular expression pattern
    fst = adms[0]
    match = re.search(pattern, fst)  # Find the matching substring
    extracted_string = match.group(1)  # Get the captured group
    path = f"{motherpath}/{fst}"
    df0 = pd.read_csv(path)
    df0['District'] = extracted_string.lower()
    exp16NDVI= df0.groupby(["Year","Month","District"]).mean().reset_index()
    # here the "district" in groupby() ensures the returned df has a District column
    for i in adms[1:]:
        match = re.search(pattern, i)  # Find the matching substring
        extracted_string = match.group(1)
        path = f"{motherpath}/{i}"
        df = pd.read_csv(path)
        df['District'] = extracted_string.lower()
        expit = df.groupby(["Year","Month","District"]).mean().reset_index()
        exp16NDVI = pd.concat([exp16NDVI,expit],axis=0)
    exp16NDVI.to_csv(return_name,index=False)
```

```python
folders = os.listdir(".")
for f in folders:
    if "." not in f:
        retname = f".Finished/{f}.csv"
        NDVIextract(f,retname)
    else: 
        continue 
```

```python
df = pd.read_csv('.Finished/NDVIAnomalies-2019.csv')

df['NDVI Anomaly lag1'] = np.nan 
df['NDVI Anomaly lag2'] = np.nan 
df['NDVI Anomaly lag3'] = np.nan 
df['NDVI Anomaly lag4'] = np.nan 
for d in set(df['District']):
    lag1 = df.query("District == @d")['NDVI Anomaly'][:-1]
    lag_1 = np.insert(lag1.values,0,np.nan)
    lag2 = df.query("District == @d")['NDVI Anomaly'][:-2]
    lag_2 = np.insert(lag2.values,[0,0],np.nan)
    lag3 = df.query("District == @d")['NDVI Anomaly'][:-3]
    lag_3 = np.insert(lag3.values,[0,0,0],np.nan)
    lag4 = df.query("District == @d")['NDVI Anomaly'][:-4]
    lag_4 = np.insert(lag4.values,[0,0,0,0],np.nan)
    tfv = df['District'] == d
    df.loc[tfv,'NDVI Anomaly lag1'] = lag_1
    df.loc[tfv,'NDVI Anomaly lag2'] = lag_2
    df.loc[tfv,'NDVI Anomaly lag3'] = lag_3
    df.loc[tfv,'NDVI Anomaly lag4'] = lag_4
    
df.to_csv('.Finished/NDVIAnomalies-2019.csv',index=False)
```
